```python
# 1. 패키지에서 모듈 불러오기
from my_math_package.calculations import basic
from my_math_package.stats import averages

def main():
    print("=== 계산기 패키지 테스트 ===")

    # 2. 기본 연산 테스트
    num1 = 10
    num2 = 5
    print(f"{num1} + {num2} = {basic.add(num1, num2)}")
    print(f"{num1} - {num2} = {basic.subtract(num1, num2)}")
    
    print("-" * 30)

    # 3. 통계 연산 및 예외 처리 테스트
    scores = [80, 90, 100]
    empty_scores = []

    # 정상적인 경우
    try:
        avg = averages.calculate_mean(scores)
        print(f"점수 리스트: {scores}")
        print(f"평균 점수: {avg}")
    except ValueError as e:
        print(f"에러 발생: {e}")

    print("-" * 30)

    # 예외가 발생하는 경우 (빈 리스트)
    print("빈 리스트로 평균 계산 시도...")
    try:
        avg = averages.calculate_mean(empty_scores)
        print(f"평균 점수: {avg}")
    except ValueError as e:
        # averages.py에서 발생시킨 ValueError를 여기서 잡아서 처리합니다.
        print(f"[오류 처리 됨] {e}")

if __name__ == "__main__":
    main()
```
